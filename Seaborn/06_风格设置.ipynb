```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 构建测试数据
def sinplot(flip=1):
    x = np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x,np.sin(x + i * 5) * (7 - i) * flip)

# matplotlib原始风格展示
sinplot()
```

```python
# 设置seaborn默认风格参数
sns.set()
sinplot()
```

#### 1. Seanborn 5种主题风格
- darkgrid
- whitegrid
- dark
- white
- ticks

##### 1.1 指定whitegrid主题

```python
sns.set_style('whitegrid')
# data = np.random.normal(size=(20,6))+np.arange(6)/2
# sns.boxplot(data = data)
sinplot()
```

##### 1.2 指定darkgrid主题

```python
sns.set_style('darkgrid')
sinplot()
```

##### 1.3 指定dark主题

```python
sns.set_style('dark')
sinplot()
```

##### 1.4 指定white主题

```python
sns.set_style('white')
sinplot()
```

##### 1.5 指定ticks主题(显示刻度)

```python
sns.set_style('ticks')
sinplot()
```

#### 2. 其他设置

##### 2.1 去除顶部和右侧的边框（despine）

```python
sinplot()
sns.despine()
```

#####  2.2 设置图与轴线的距离（despine()的offset参数）

```python
sinplot()
sns.despine(offset=30)
```

##### 2.3 轴线隐藏（despine()的left,,top,right,bottom）

```python
sns.set_style('ticks')
sinplot()
sns.despine(left=True)
```

```python
sns.set_style('ticks')
sinplot()
sns.despine(bottom=True)
```

#### 3. 指定子图不同风格 with sns.axes_style('ticks') 

```python
with sns.axes_style('ticks'):
    plt.subplot(211)
    sinplot()

plt.subplot(212)
sinplot(-1)
```

```python
sns.set()
```

####  4. set_context

##### 4.1 设置线的粗细(set_context参数)
- paper
- talk
- poster
- notebook

或者修改rc参数
>rc={"lines.linewidth":3}

```python
sns.set_context('paper')
# 设置面板大小
plt.figure(figsize=(8,6))
sinplot()
```

```python
sns.set_context('talk')
plt.figure(figsize=(8,6))
sinplot()
```

```python
sns.set_context('poster')
plt.figure(figsize=(8,6))
sinplot()
```

```python
sns.set_context('poster',rc={"lines.linewidth":10})
sinplot()
```

#####  4.2 设置字体大小

```python
sns.set_context('talk',font_scale=2.5)
sinplot()
```
