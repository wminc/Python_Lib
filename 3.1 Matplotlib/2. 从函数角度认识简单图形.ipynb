```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

# 指定默认字体
plt.rcParams['font.sans-serif'] = ['SimHei']
# 解决保存图像的'-'显示方块的问题
plt.rcParams['axes.unicode_minus'] = False
```

#### 1. bar()：用于绘制柱状图
> plt.bar(x, y)

**参数说明：**
- x：x轴上的数据类别
- y：x轴上的各数据类别对应的数量
- align：位置
- color：柱的颜色
- tick_label：刻度标签，默认是x的值
- hatch：柱上的图形

```python
x = [1,2,3,4,5,6,7,8]
y = [3,1,4,5,8,9,7,2]

#plt.bar(x,y,align="center",color="c",tick_label=["a","b","c","d","e","f","g","h"],hatch="/")
plt.bar(x,y,align="center",color="c",hatch="/")

plt.xlabel("箱子编号")
plt.ylabel("箱子重量(kg)")

plt.show()
```

#### 2. barh()：用于绘制条形图
> plt.barh(x, y)

**参数说明：**
- x：x轴上的数据类别
- y：x轴上的各数据类别对应的数量
- align：位置
- color：柱的颜色
- tick_label：刻度标签，默认是x的值
- hatch：柱上的图形

```python
x = [1,2,3,4,5,6,7,8]
y = [3,1,4,5,8,9,7,2]

#plt.bar(x,y,align="center",color="c",tick_label=["a","b","c","d","e","f","g","h"],hatch="/")
plt.barh(x,y,align="center",color="y",hatch="X")

plt.xlabel("箱子编号")
plt.ylabel("箱子重量(kg)")

plt.show()
```

#### 3. hist()：用于绘制直方图
> plt.hist(x)

**参数说明：**
- x：x轴上的数据类别
- bins：
- histtype：类型
- rwidth：
- alpha：透明度

```python
boxWeight = np.random.randint(0,10,100)
x = boxWeight

bins = range(0,11,1)

#plt.bar(x,y,align="center",color="c",tick_label=["a","b","c","d","e","f","g","h"],hatch="/")
plt.hist(x,bins=bins,color="g",histtype="bar",rwidth=1,alpha=0.6)

plt.xlabel("箱子重量(kg)")
plt.ylabel("箱子重量(个)")

plt.show()
```

#### 4. pie()：用于绘制饼图（绘制定性数据的不同类别的百分比）
> plt.pie(x)

**参数说明：**
- x：定性数据的不同类别的百分比
- labels：种类
- autopct：占比
- startangle：
- colors：颜色

```python
kinds = "简易箱","保温箱","行李箱","密封箱"
colors = ["#e41a1c","#377eb8","#4daf4a","#984ea3"]
soldNums = [0.05,0.45,0.15,0.35]

plt.pie(soldNums,labels=kinds,autopct="%3.lf%%",startangle=60,colors=colors)
plt.title("不同类型箱子的销售数量占比")
plt.show()
```

#### 5. polar()：用于绘制极线图（在极坐标轴上绘制折线图）
> plt.polar(theta,r)

**参数说明：**
- theta：每个标记所在射线与极径的夹角
- r：每个标记到原点的距离
- color：颜色
- lw：线宽
- marker：标记点的样式
- mfc：
- ms：

```python
barSlices = 12

theta = np.linspace(0.0,2*np.pi,barSlices,endpoint=False)
r = 30* np.random.rand(barSlices)

plt.polar(theta,r,color="chartreuse",lw=2,marker="*",mfc="b",ms=10)

plt.show()
```

#### 6. scatter()：用于绘制气泡图
> plt.scatter(x,y)

**参数说明：**
- x：x轴上的数值
- y：x轴上的数值
- c：散点中标记的颜色
- s：散点标记的大小
- label：标记图形内容的标签文本

```python
x = np.linspace(0.05, 10, 1000)
y = np.random.rand(1000)

plt.scatter(x,y,c="g",label="scatter figure")
plt.legend()
plt.show()
```

```python
a = np.random.randn(100)
b = np.random.randn(100)

plt.scatter(a,b,s=np.power(10*a+20*b,2),c=np.random.rand(100),cmap=mpl.cm.RdYlBu,marker="o")
plt.show()
```

#### 7. stem()：用于绘制棉棒图
> plt.stem(x,y)

**参数说明：**
- x：指定棉棒x轴基线上的位置
- y：绘制棉棒的长度
- linefmt：棉棒样式
- markerfmt：棉棒末端的样式
- basefmt：指定基线样式

```python
x = np.linspace(0.5,2*np.pi,20)
y = np.random.randn(20)

plt.stem(x,y,linefmt="-.",markerfmt="o",basefmt="-")
plt.show()
```

#### 8. boxplot()：用于绘制箱线图
> plt.boxplot(x)

**参数说明：**
- x：绘制箱线图输入的数据

```python
x = np.random.randn(1000)

plt.boxplot(x)

#plt.xlabel("随机数生成器AlphaRM")
plt.xticks([1],["随机数生成器AlphaRM"])

plt.ylabel("随机数值")
plt.title("随机数生成器抗干扰能力的稳定性")

# 只绘制垂直于y轴的网格线
plt.grid(axis="y",ls=":",lw=1,color="gray",alpha=0.4)
plt.show()
```

#### 9. errorbar()：用于绘制误差棒图
> plt.errorbar(x,y,yerr=a,err=b)

**参数说明：**
- x：数据点水平位置
- y：数据点垂直位置
- yerr：y轴方向的数据点的误差值
- err：x轴方向的数据点的误差值

```python
x = np.linspace(0.1,0.6,6)
y = np.exp(x)

plt.errorbar(x,y,fmt="bo:",yerr=0.2,xerr=0.02)
plt.xlim(0,0.7)
plt.show()
```
