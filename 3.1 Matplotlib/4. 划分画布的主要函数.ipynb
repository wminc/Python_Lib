```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

# 指定默认字体
plt.rcParams['font.sans-serif'] = ['SimHei']
# 解决保存图像的'-'显示方块的问题
plt.rcParams['axes.unicode_minus'] = False
```

### 1. 函数subplot()：绘制网格区域中的几何形状相同的子区布局
> 假设subplot()的三个参数分别是整数C,R,P，即subplot(C,R,P),那么就表示在C行，R列的网格布局中，子区放置在第P个位置上（子区编号始于左上角，序号依次享有递增）,通过调用函数subplot获得坐标轴实例ax,我们可以用ax添加图形
#### 1.1 函数subplot()的适用方法

```python
x = np.linspace(-2*np.pi,2*np.pi,200)
y = np.sin(x)
y1 = np.cos(x)

plt.subplot(121)
plt.plot(x,y)
plt.subplot(122)
plt.plot(x,y1)
plt.show()
```

#### 1.2 案例——在极坐标轴上绘制折线图

```python
radii = np.linspace(0,1,100)
theta = 2*np.pi*radii
ax = plt.subplot(111,polar=True)

ax.plot(theta,radii,color='r',linestyle='-',linewidth=2)
plt.show()
```

#### 1.3 案例——在极坐标轴上绘制散点图

```python
radii = 30*np.random.rand(100)
theta = 2*np.pi*np.random.rand(100)
colors = np.random.rand(100)
size = 50*radii
ax = plt.subplot(111,polar=True)

ax.scatter(theta,radii,s=size,c=colors,cmap=mpl.cm.PuOr,marker="*")
plt.show()
```

#### 1.4 案例——在非等分画布的绘图区域上实现图形展示
**注意：**这里需要创建画布，画布是为了去添加多个坐标轴

```python
# 创建画布
fig = plt.figure()

x = np.linspace(0.0,2*np.pi)
y = np.cos(x)*np.sin(x)

# 创建坐标轴
ax1 = fig.add_subplot(121)
ax1.margins(0.03)
ax1.plot(x,y,ls="-",lw=2,color="b")

# 创建坐标轴
ax2 = fig.add_subplot(222)
ax2.margins(0.7,0.7)
ax2.plot(x,y,ls="-",lw=2,color="r")

# 创建坐标轴
ax3 = fig.add_subplot(224)
ax3.margins(0.1,0.3)
ax3.plot(x,y,ls="-",lw=2,color="g")

plt.show()
```

### 2. 函数subplot2grid()：让子区域跨越固定的网格布局
>subplot()只能绘制等分画布形式的图形样式，进行费等分画布形式的图形展示时，需要像话不多次使用subplot()例如1.4案例 这样显得很麻烦，而且容易出现差错。因此我们需要其他方法实现定制化的网格区域——subplot2grid()

#### 2.1 函数subplot2grid()的使用

```python
plt.subplot2grid((2,3),(0,0),colspan=2)
x = np.linspace(0.0,4.0,100)
y = np.random.randn(100)
plt.scatter(x,y,c="c")
plt.title("散点图")

plt.subplot2grid((2,3),(0,2))
plt.title("空白区域图")

plt.subplot2grid((2,3),(1,0),colspan=3)
x = np.linspace(0.0,4.0,100)
y1 = np.sin(x)
plt.plot(x,y1,lw=2,ls="-")
plt.xlim(0,3)
plt.grid(True,ls=":",c="r")
plt.title("折线图")

# 设置画布标题
plt.suptitle("subplot2grid()函数展示",fontsize=15)
# 设置子区域填充整个画布，但会影响画布标题的显示
#plt.tight_layout()

#调整边距和子图的间距
#plt.subplots_adjust(left=None, bottom=None, right=None, top=None,
#                wspace=None, hspace=None)

plt.subplots_adjust(wspace=0.5,hspace=0.5)
plt.show()
```

### 3. 函数subplots()：类似于subplot2grid()的功能
>函数subplots()，返回值是一个元组(fig,ax),其中fig是画布Figure实例，ax是坐标轴axis对象，如果创建多子区，则ax是一个数组

```python
x = np.linspace(0,2*np.pi,500)
y = np.sin(x)*np.exp(-x)

fig,ax = plt.subplots(1,2,sharey=True)

ax1 = ax[0]
ax1.plot(x,y,"k--",lw=2)
ax1.grid(ls=':',lw=1,color="gray",alpha=0.8)
ax1.set_title('折线图')

ax2 = ax[1]
ax2.scatter(x,y,s=10,c="skyblue",marker="o")
ax2.set_title("散点图")

plt.suptitle("subplots()函数使用")

plt.show()
```

### 扩展：subplots_adjust()改变子区相距画布边缘的距离和自取边缘之间的高度与宽度距离
>subplots_adjust(left=None, bottom=None, right=None, top=None,wspace=None, hspace=None)

- wspace：控制子区之间的宽度距离
- hspace：控制子区之间的高度距离
