```python
import numpy as np
import matplotlib.pyplot as plt
```

#### 1. plot()：展示变量趋势变化
> plt.plot(x, y, ls="-", lw=2, label="plot figure")

**参数说明：**
- x：x轴上的数值
- y：y轴上的数值
- ls：折线图线条风格
- lw：折线图线条宽度
- label：标记图形内容的标签文本

```python
x = np.linspace(0.05, 10, 1000)
y = np.cos(x)

plt.plot(x,y,ls="-",lw=2,label="plot figure")
plt.legend()
plt.show()
```

#### 2. scatter()：寻找变量之间关系
>plt.scatter(x, y, c="b", label="scatter figure")

**参数说明：**
- x：x轴上的数值
- y：x轴上的数值
- c：散点图中标记的颜色
- label：标记图形内容的标签文本


```python
x = np.linspace(0.05, 10, 1000)
y = np.random.rand(1000)

plt.scatter(x,y,c="g",label="scatter figure")
plt.legend()
plt.show()
```

#### 3. xlim()：设置x轴数值范围
>plt.xlim(xmin, xmax)

**参数说明：**
- xmin：x轴上的最小值
- xmax：x轴上的最大值

上面函数功能，调用签名和参数说明同样可以品一道函数**ylim()**上


```python
x = np.linspace(0.05, 10, 1000)
y = np.random.rand(1000)

plt.scatter(x,y,c="g",label="scatter figure")
plt.legend()

plt.xlim(0.05, 10)
plt.ylim(0, 1)
plt.show()
```

#### 4. xlabel()：设置x轴的标签文本
>plt.xlabel(string)

**参数说明：**
- string：标签文本内容

上面函数功能，调用签名和参数说明同样可以品一道函数**ylabel()**上

```python
x = np.linspace(0.05, 10,1000)
y = np.sin(x)

plt.plot(x, y, ls='-.', lw=2, c='c', label='plot figure')

plt.legend()

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

#### 5. grid()：绘制刻度线的网格线
>plt.grid(linestyle=", color="”)

**参数说明：**
- linestyle：网格线的线条风格
- color：网格线的线条颜色

```python
x = np.linspace(0.05, 10, 1000)
y = np.sin(x)

plt.plot(x,y,ls='-.',lw=2,c='c',label='plot figure')
plt.legend()

plt.grid(linestyle=':',color='r')
plt.show()
```

#### 6. axhline()：绘制平行于x轴的水平参考线
>plt.axhline(y=0.0, c="r",ls="--",lw=2)

**参考说明：**
- y：水平参考线的值
- c：参考线的线条颜色
- ls：参考线的线条风格
- lw：参考线的线条宽度

上述函数功能，调用签名和参数说明可以平移到函数**axvline()**上

```python
x = np.linspace(0.05, 10, 1000)
y = np.sin(x)

plt.plot(x,y,ls='-.',lw=2,c='c',label='plot figure')
plt.legend()

plt.axhline(y=0.0,c='r',ls='--',lw=2)
plt.axvline(x=4.0,c='r',ls='--',lw=2)
plt.show()
```

#### 7. axvspan()：绘制垂直于x轴的参考区域
>plt.axvspan(xmin=1.0,xmax=2.0,facecolor='y',alpha=0.3)

**参数说明：**
- xmin：参考区域的起始位置
- xmax：参考区域的终止位置
- facecolor：参考区域的填充颜色
- alpha：参考区域的填充颜色的透明度

上述函数功能，调用签名和参数说明可以平移到函数axhspan()上

```python
x = np.linspace(0.05, 10, 1000)
y = np.sin(x)

plt.plot(x,y,ls='-.',lw=2,c='c',label='plot figure')
plt.legend()

plt.axhspan(ymin=0.0,ymax=0.5,facecolor='y',alpha=0.3)
plt.axvspan(xmin=4.0,xmax=6.0,facecolor='y',alpha=0.3)
plt.show()
```

#### 8. annotate()：添加图形内容细节的指向型注释文本
>plt.annotate(string,xy=(np.pi/2,1.0),xytext=((np.pi/2)+0.15,1.5),wight="bold",color="b",arrowprops=dict(arrowstyle="->",connectionstyle="arc3",color="b"))

**参数说明：**
- string：图形内容的注释文本
- xy：被注释图形内容的位置坐标
- xytext：注释文本的位置坐标
- weight：注释文本的字体粗细风格
- color：注释文本的字体颜色
- arrowprops：指示被注释内容的箭头的属性字典


```python
x = np.linspace(0.5,10,1000)
y = np.sin(x)

plt.plot(x,y,ls='-.',lw=2,c='c',label='plot figure')
plt.legend()

plt.annotate("maximum",xy=(np.pi/2,1.0),xytext=((np.pi/2)+1.0,.9),weight="bold",color="b",arrowprops=dict(arrowstyle="->",connectionstyle="arc3",color="b"))

plt.show()
```

#### 9. text()：添加图形内容细节的无指向型注释文本
>plt.text(x, y, string, weight="bold",color="b")

**参考说明：**
- x：注释文本内容所在位置的横坐标
- y：注释文本内容所在位置的纵坐标
- string：注释文本内容
- weight：注释文本内容的粗细风格
- color：注释文本内容的字体颜色

```python
x = np.linspace(0.5,10,1000)
y = np.sin(x)

plt.plot(x,y,ls='-.',lw=2,c='c',label='plot figure')
plt.legend()

plt.text(3.10,0.09,"y=sin(x)",weight="bold",color="b")

plt.show()
```

#### 10. title()：添加图形内容的标题
>plt.title(string)

**参数说明：**
- string：图形内容的标题文本

```python
x = np.linspace(0.5,10,1000)
y = np.sin(x)

plt.plot(x,y,ls='-.',lw=2,c='c',label='plot figure')
plt.legend()

plt.title("y=sin(x)")

plt.show()
```

#### 11. legend()：标示不同图形的文本标签图例
>plt.legend(loc="lower left")

**参数说明：**
- loc：图例在图中的地理位置
    >best,upper right,upper left,lower left,lower right,right,
    center left,center right,lower center,upper center,center

```python
x = np.linspace(0.5,10,1000)
y = np.sin(x)

plt.plot(x,y,ls='-.',lw=2,c='c',label='plot figure')

plt.legend(loc="best")
# plt.legend(loc="center")
# plt.legend(loc="upper right")
# plt.legend(loc="lower right")
# plt.legend(loc="lower left")

plt.show()
```
