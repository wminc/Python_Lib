```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cm as sm
```

```python
# 定义数据
x = np.linspace(0.5, 3.5, 100)
y = np.sin(x)
y1 = np.random.randn(100)
```

```python
# scatter散点图
plt.scatter(x,y1,c="0.25",label='scatter figure')

# plot曲线图
plt.plot(x,y,ls="--",lw=2,label="plot figure")

for spine in plt.gca().spines.keys():
    if spine == "top" or spine == "right":
        plt.gca().spines[spine].set_color("none")

plt.gca().xaxis.set_ticks_position("bottom")
plt.gca().yaxis.set_ticks_position("left")

# 设置x,y轴范围
plt.xlim(0.0,4.0)
plt.ylim(-3.0,3.0)

# 设置x,y轴标签
plt.xlabel("x_axis")
plt.ylabel("y_axis")

# 设置网格线
plt.grid(True,ls=":",color="r")

# 设置平行于x轴的参考线
plt.axhline(y=0.0,c="r",ls="--",lw=2)

# 设置垂直x轴的参考区域
plt.axvspan(xmin=1.0,xmax=2.0,facecolor='y',alpha=.3)

# 设置内容细节，指向型注释文本
plt.annotate("maximum",xy=(np.pi/2,1.0),xytext=((np.pi/2)+0.15,1.5),weight="bold",color="r",arrowprops=dict(arrowstyle="->",connectionstyle="arc3",color="r"))

plt.annotate("spines",xy=(0.75,-3),xytext=(0.35,-2.25),weight="bold",color="b",
             arrowprops=dict(arrowstyle="->",connectionstyle="arc3",color="b"))

plt.annotate("",xy=(0,-2.78),xytext=(0.4,-2.32),
             arrowprops=dict(arrowstyle="->",connectionstyle="arc3",color="b"))

# 箭头指向下面的无指向性文本
#plt.annotate("",xy=(3.5,-2.98),xytext=(3.6,-2.70),
#             arrowprops=dict(arrowstyle="->",connectionstyle="arc3",color="b"))

# 设置内容细节，无指向型注释文本
plt.text(3.6,-2.60,"'|' is tickline",weight="bold",color="b")
plt.text(3.6,-2.95,"3.5 is ticklabel",weight="bold",color="b")

# 设置图形内容标题
plt.title("structure of matplotlib")

# 设置图例
plt.legend()

plt.show()
```
