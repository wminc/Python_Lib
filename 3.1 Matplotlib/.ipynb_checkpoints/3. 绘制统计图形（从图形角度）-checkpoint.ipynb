```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

# 指定默认字体
plt.rcParams['font.sans-serif'] = ['SimHei']
# 解决保存图像的'-'显示方块的问题
plt.rcParams['axes.unicode_minus'] = False
```

### 1. 柱状图
>bar(x,y,align="center",color="b",tick_label=["A","B","C","D","E"],alpha=0.6)

- x：柱状图中的柱体标签
- y：柱状图中的柱体高度
- align：主体对其方式
- color：柱体颜色
- tick_label：刻度标签值
- alpha：柱体透明度


```python
x = [1,2,3,4,5]
y = [6,10,4,5,1]

plt.bar(x,y,align="center",color="b",tick_label=["A","B","C","D","E"],alpha=0.6)

plt.xlabel('测试难度')
plt.ylabel('试卷份数')

plt.grid(True,axis="y",ls=":",color="r",alpha=0.3)
plt.show()
```

### 2. 条形图
>barh(x,y,align="center",color="b",tick_label=["A","B","C","D","E"])

```python
x = [1,2,3,4,5]
y = [6,10,4,5,1]

plt.barh(x,y,align="center",color="b",tick_label=["A","B","C","D","E"],alpha=0.6)

plt.xlabel('测试难度')
plt.ylabel('试卷份数')

plt.grid(True,axis="x",ls=":",color="r",alpha=0.3)
plt.show()
```

### 3. 堆积图
#### 3.1 堆积柱状图
>bar(x,y1,bottom=y,color='')

堆积柱状图相当于在某一柱状图基础上堆积一个柱状图。将函数bar()中的参数bottom取值设定为y,就会输出堆积柱状图



```python
x = [1,2,3,4,5]
y = [6,10,4,5,1]
y1 = [2,6,3,8,5]

plt.bar(x,y,align="center",color="#66c2a5",tick_label=["A","B","C","D","E"],label="班级A")
plt.bar(x,y1,align="center",bottom=y,color="#8da0cb",label="班级B")

plt.xlabel('测试难度')
plt.ylabel('试卷份数')

plt.legend()
plt.show()
```

#### 3.1 堆积条形图
>barh(x,y1,left=y,color='')

堆积柱状图相当于在某一条形图基础上堆积一个条形图。将函数bar()中的参数left取值设定为y,就会输出堆积条形图

```python
x = [1,2,3,4,5]
y = [6,10,4,5,1]
y1 = [2,6,3,8,5]

plt.barh(x,y,align="center",color="#66c2a5",tick_label=["A","B","C","D","E"],label="班级A")
plt.barh(x,y1,align="center",left=y,color="#8da0cb",label="班级B")

plt.xlabel('测试难度')
plt.ylabel('试卷份数')

plt.legend()
plt.show()
```

### 4. 分块图
#### 4.1 多数据并列柱状图
>bar(x+bar_width,y1)

```python
x = np.arange(5)
print(x)
y = [6,10,4,5,1]
y1 = [2,6,3,8,5]

bar_width = 0.35
tick_label=["A","B","C","D","E"]

plt.bar(x,y,bar_width,align="center",color="#66c2a5",label="班级A")
plt.bar(x+bar_width,y1,bar_width,align="center",color="#8da0cb",label="班级B")

plt.xlabel('测试难度')
plt.ylabel('试卷份数')

# 设置刻度在两个柱状图之间
plt.xticks(x+bar_width/2,tick_label)

plt.legend()
plt.show()
```

#### 4.2 多数据并列条形图
>barh(x+bar_width,y1)

```python
x = np.arange(5)
print(x)
y = [6,10,4,5,1]
y1 = [2,6,3,8,5]

bar_width = 0.35
tick_label=["A","B","C","D","E"]

plt.barh(x,y,bar_width,align="center",color="#66c2a5",label="班级A")
plt.barh(x+bar_width,y1,bar_width,align="center",color="#8da0cb",label="班级B")

plt.xlabel('测试难度')
plt.ylabel('试卷份数')

# 设置刻度在两个柱状图之间
plt.yticks(x+bar_width/2,tick_label)

plt.legend()
plt.show()
```

### 6. 堆积折线图、间断条形图、阶梯图
#### 6.1 stackplot()绘制堆积折线图

```python
x = np.arange(1,6,1)
#print(x)
y = [0,4,3,5,6]
y1 = [1,3,4,2,7]
y2 = [3,4,1,6,5]

labels=["BluePlanet","BrownPlanet","GreenPlanet"]
colors = ["#8da0cb","#fc8d62","#66c2a5"]

plt.stackplot(x,y,y1,y2,labels=labels,colors=colors)

plt.legend(loc="upper left")
plt.show()
```

#### 6.2 broken_barh()绘制间断条形图

```python
plt.broken_barh([(30,100),(180,50),(260,70)],(20,8),facecolor="#1f78b4")
# 列表表示从起点是x轴的数值为60的位置向x轴正方向移动90个单位
# (10,8)表示从起点是y轴的数值为10的位置起，延y轴正方向移动8个单位，这就是柱体的高度和起始位置
plt.broken_barh([(60,90),(190,20),(230,30),(280,60)],(10,8),
                 facecolor=("#7fc97f","#beaed4","#fdc086","#ffff99"))

plt.xlim(0,360)
plt.ylim(5,35)
plt.xlabel("演出时间")

plt.xticks(np.arange(0,361,60))
plt.yticks([15,25],["歌剧院A","歌剧院B"])
plt.grid(ls="-",lw=1,color="gray")

plt.title("不同地区上午歌剧院的时间比较")
plt.show()
```

#### 6.3 step()绘制间断阶梯图
>where参数默认是"pre",x轴上相邻数据点的取值是左开右闭选取，还可以取post左闭右开

```python
x = np.linspace(1,10,10)
print(x)
y = np.sin(x)

plt.step(x,y,color="#8dd3c7",where="pre",lw=2)

plt.xlim(0,11)
plt.xticks(np.arange(1.11,1))
plt.ylim(-1.2,1.2)
plt.show()
```

### 7. 直方图
>hist(x,bins=bins,color="#377eb8",histtype="bar",rwidth=1.0)

- x：连续型数据输入值
- bins：柱体个数
- color：柱体颜色
- histtype：柱体类型
- label：图例内容
- rwidth：柱体宽度，取值[0,1]

```python
scoresT = np.random.randint(0,100,100)

x = scoresT

bins = range(0,101,10)

plt.hist(x,bins=bins,color="#377eb8",histtype="bar",rwidth=1.0)

plt.xlabel("测试成绩")
plt.ylabel("学生人数")

plt.show()
```

#### 7.1 直方图和柱状图的关系
- 直方图和柱状图很相似，但是直方图面熟的是连续型数据的分布，柱状图描述的是离散型数据的分布
- 柱状图的主体之间有空隙，直方图的柱体之间没有

#### 7.2 堆积直方图
stacked=True展示堆积效果，stacked=False并排展示

```python
scoresT1 = np.random.randint(0,100,100)
scoresT2 = np.random.randint(0,100,100)
print(scoresT1)
x = [scoresT1,scoresT2]

colors = ["#8dd3c7","#bebada"]
labels = ["班级A","班级B"]

bins = range(0,101,10)

plt.hist(x,bins=bins,color=colors,histtype="bar",rwidth=1.0,stacked=True,label=labels)

plt.xlabel("测试成绩")
plt.ylabel("学生人数")

plt.title("不同班级的测试成绩直方图")
plt.legend(loc="upper left")
plt.show()
```

#### 7.3 直方图的不同形状¶

```python
scoresT1 = np.random.randint(0,100,100)
scoresT2 = np.random.randint(0,100,100)
print(scoresT1)
x = [scoresT1,scoresT2]

colors = ["#8dd3c7","#bebada"]
labels = ["班级A","班级B"]

bins = range(0,101,10)

plt.hist(x,bins=bins,color=colors,histtype="stepfilled",rwidth=1.0,stacked=True,label=labels)

plt.xlabel("测试成绩")
plt.ylabel("学生人数")

plt.title("不同班级的测试成绩直方图")
plt.legend(loc="upper left")
plt.show()
```

### 8. 饼图
> pie(percent,explode,labels,autopct,startangle,shadow,colors)

- percent：百分比
- explode：饼片边缘偏离半径的百分比
- labels：标记每份饼片的文本标签信息
- autopct：饼片文本标签内容对应的数值百分比样式
- startangle：从x轴作为起始位置，第一个饼片逆时针旋转的角度
- shadow：是否会之饼片的阴影
- colors：饼片的颜色

#### 8.1 分裂式饼图

```python
labels = "A难度水平","B难度水平","C难度水平","D难度水平"
percent = [0.35,0.15,0.20,0.30]
colors = ["#377eb8","#4daf4a","#984ea3","#ff7f00"]
explode = (0.1,0.1,0.1,0.1)

plt.pie(percent,explode=explode,
        labels=labels,
        autopct="%3.lf%%",
        startangle=45,
        shadow=True,
        colors=colors
       )
plt.title("选择不同难度测试试卷的学生占比")
plt.show()
```

#### 8.2 非分裂式饼图

```python
labels = "A难度水平","B难度水平","C难度水平","D难度水平"
percent = [0.35,0.15,0.20,0.30]
colors = ["#377eb8","#4daf4a","#984ea3","#ff7f00"]

plt.pie(percent,
        labels=labels,
        autopct="%3.lf%%",
        startangle=45,
        pctdistance=0.7,
        labeldistance=1.2,
        colors=colors
       )
plt.title("选择不同难度测试试卷的学生占比")
plt.show()
```

#### 8.3 内嵌环形饼图

```python
elements = ["面粉","砂糖","奶油","草莓酱","坚果"]
weight1 = [40,15,20,10,15]
weight2 = [30,25,15,20,10]
colormaplist = ["#e41a1c","#377eb8","#4daf4a","#984ea3","#ff7f00"]
outer_color = colormaplist
inner_color = colormaplist

wedges1,texts1,autotexts1 = plt.pie(weight1,
                                  autopct="%3.lf%%",
                                  radius=1,
                                  pctdistance=0.85,
                                  colors=outer_color,
                                  textprops=dict(color="w"),
                                  wedgeprops=dict(width=0.3,edgecolor="w"))

wedges2,texts2,autotexts2 = plt.pie(weight2,
                                  autopct="%3.lf%%",
                                  radius=0.7,
                                  pctdistance=0.75,
                                  colors=inner_color,
                                  textprops=dict(color="w"),
                                  wedgeprops=dict(width=0.3,edgecolor="w"))

plt.legend(wedges1,elements,fontsize=12,
          title="配料表",
          loc="center left",
          bbox_to_anchor=(0.91,0,0.3,1))
plt.setp(autotexts1,size=13,weight="bold")
plt.setp(autotexts2,size=13,weight="bold",)
plt.setp(texts1,size=12)

plt.title("不同果酱面包配料表")
plt.show()
```

### 9. 箱线图
>boxplot(testList,whis=whis,widths=width,sym="o",labels=labels,patch_artist=True)

- testList：输入数据
- whis：四分位间距的倍数，用来确定想须包含数据的范围大小
- widths：设置箱体的宽度
- sym：离群值得标记样式
- labels：绘制数据集的刻度标签
- patch_artist：是否给箱体添加颜色

#### 9.1 离群值的垂直箱线图

```python
testA = np.random.randn(5000)
testB = np.random.randn(5000)

testList = [testA,testB]
labels = ["随机数生成器AlphaRM","随机数生成器BetaRM"]
colors = ["#1b9e77","#d95f02"]

whis = 1.6
width = 0.35

bplot = plt.boxplot(testList,whis=whis,widths=width,sym="o",labels=labels,patch_artist=True)

for patch,color in zip(bplot["boxes"],colors):
    patch.set_facecolor(color)
    
plt.ylabel("随机数量")
plt.title("生成器抗干扰能力稳定性比较")

plt.grid(axis="y",ls=":",lw=1,color="gray",alpha=0.4)

plt.show()
```

#### 9.2 离群值的水平箱线图

```python
x = np.random.randn(1000)

bplot = plt.boxplot(x,vert=False)
    
plt.xlabel("随机数值")
plt.yticks([1],["随机数生成器AlphaRM"],rotation=90)
plt.title("生成器抗干扰能力稳定性比较")

plt.grid(axis="y",ls=":",lw=1,color="gray",alpha=0.4)

plt.show()
```

#### 9.3 不绘制离群值的水平箱线图

```python
x = np.random.randn(1000)

bplot = plt.boxplot(x,vert=False,showfliers=False)
    
plt.xlabel("随机数值")
plt.yticks([1],["随机数生成器AlphaRM"],rotation=90)
plt.title("生成器抗干扰能力稳定性比较")

plt.grid(axis="y",ls=":",lw=1,color="gray",alpha=0.4)

plt.show()
```

### 10. 误差棒图
>errorbar(x,y,yerr=error_limit,fmt=":o",ms=5,mfc="c",mec="r",capthick=1,capsize=2)

- x,y：数据点的位置
- yerr：单一数值的非对称形式误差范围
- fmt：数据点的标记样式和数据点标记的连接线的样式
- ecolor：误差棒的线条颜色
- elinewidth：误差棒的线条粗细
- ms：数据点的大小
- mfc：数据点的标记颜色
- mec：数据点的标记边缘的颜色
- capthick：误差棒边界横杠的厚度
- capsize：误差棒边界横杠的大小

```python
x = np.linspace(0.1,0.6,10)
y = np.exp(x)

error = 0.05+0.15*x

lower_error = error
upper_error = 0.3*error
error_limit = [lower_error,upper_error]

plt.errorbar(x,y,yerr=error_limit,
            fmt=":o",
            ms=5,
            mfc="c",
            mec="r",
            capthick=1,capsize=2)
plt.xlim(0,0.7)
plt.show()
```

#### 10.1 带误差棒的柱状图

```python
x = np.arange(5)
y = [100,68,79,91,82]
std_err = [7,2,6,10,5]
error_attri = dict(elinewidth=2,ecolor="black",capsize=3)

plt.bar(x,y,color="c",width=0.6,align="center",yerr=std_err,error_kw=error_attri,
       tick_label=["园区1","园区2","园区3","园区4","园区5"])

plt.xlabel("芒果种植区")
plt.ylabel("收割量")

plt.title("不同芒果种植区的单次收割量")

plt.grid(True,axis="y",ls=":",color="gray",alpha=0.2)

plt.show()
```

#### 10.2 带误差棒的条形图

```python
x = np.arange(5)
y = [1200,2400,1800,2200,1600]
std_err = [150,100,180,130,80]
bar_width = 0.6
colors = ["#e41a1c","#377eb8","#4daf4a","#984ea3","#ff7f00"]

plt.barh(x,y,bar_width,
          color=colors,
          align="center",
          xerr=std_err,
          tick_label=["家庭","小说","心理","科技","儿童"])

plt.xlabel("订购数量")
plt.ylabel("图书种类")

plt.title("大型图书展销会的不同图书种类的采购情况")

plt.grid(True,axis="x",ls=":",color="gray",alpha=0.2)
plt.xlim(0,2600)
plt.show()
```

#### 10.3 带误差棒的多数据并列柱状图

```python
x = np.arange(5)
y1 = [100,68,79,91,82]
y2 = [120,75,70,78,85]
std_err1 = [7,2,6,10,5]
std_err2 = [5,1,4,8,9]

error_attri = dict(elinewidth=2,ecolor="black",capsize=3)

bar_width=0.4
tick_label = ["园区1","园区2","园区3","园区4","园区5"]
plt.bar(x,y1,color="#87CEEB",width=bar_width,align="center",
        yerr=std_err1,
        error_kw=error_attri,
       label="2010")
plt.bar(x+bar_width,y2,color="#CD5C5C",width=bar_width,align="center",
        yerr=std_err2,
        error_kw=error_attri,
       label="2013")

plt.xlabel("芒果种植区")
plt.ylabel("收割量")
plt.xticks(x+bar_width/2,tick_label)
plt.title("不同芒果种植区的单次收割量")

plt.grid(True,axis="y",ls=":",color="gray",alpha=0.2)
plt.legend()
plt.show()
```

#### 10.4 带误差棒的堆积柱状图

```python
x = np.arange(5)
y1 = [1200,2400,1800,2200,1600]
y2 = [1050,2100,1300,1600,1340]
std_err1 = [150,100,180,130,80]
std_err2 = [120,110,170,150,120]
bar_width = 0.6
tick_label=["家庭","小说","心理","科技","儿童"]

error_attri = dict(elinewidth=2,ecolor="black",capsize=0)

#colors = ["#e41a1c","#377eb8","#4daf4a","#984ea3","#ff7f00"]

plt.bar(x,y1,bar_width,
          color="#6495ED",
          align="center",
          yerr=std_err1,
        label="地区1",
        error_kw=error_attri
          )
plt.bar(x,y2,bar_width,
          color="#FFA500",
          align="center",
          yerr=std_err2,
        label="地区2",
        error_kw=error_attri
          )

plt.ylabel("订购数量")
plt.xlabel("图书种类")

plt.title("大型图书展销会的不同图书种类的采购情况")

plt.grid(True,axis="y",ls=":",color="gray",alpha=0.2)
plt.xticks(x,tick_label)
plt.legend()
plt.show()
```
