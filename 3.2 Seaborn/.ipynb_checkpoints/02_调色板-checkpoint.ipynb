```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

### 调色板
- color_palette()能传入任何Matplotlib所支持的颜色
- color_palette()不写参数则默认
- set_palette()设置所有图的颜色

#### 1. 分类色板

```python
current_palette = sns.color_palette()
sns.palplot(current_palette)
```

#### 2. 圆形色板(hls)
##### 2.1 类似于hls色彩空间,设定分成的份数

```python
sns.palplot(sns.color_palette('hls',10))
```

```python
data = np.random.normal(size=(20,8))+np.arange(8)/2
sns.boxplot(data=data,palette=sns.color_palette('hls',8))
```

##### 2.2 hls_palette()来控制颜色的亮度和饱和度
- l:亮度(lightness)
- s:饱和度(saturation)

```python
sns.palplot(sns.hls_palette(8, l=.3, s=.8))
```

##### 2.3 Paired色彩不同一级分类下相同二级分类使用对比色

```python
sns.palplot(sns.color_palette('Paired',10))
```

#### 3. 使用xkcd颜色来命名颜色
>xkcd包含了针对随机RGB颜色的命名，产生954个随时通过xkcd_rgb字典中调用的命名颜色

```python
plt.plot([0,1],[0,1],sns.xkcd_rgb["pale red"],lw=3)
plt.plot([0,1],[0,2],sns.xkcd_rgb["medium green"],lw=3)
plt.plot([0,1],[0,3],sns.xkcd_rgb["denim blue"],lw=3)
```

#### 4. 连续色板
##### 4.1 色彩随数据变化，数据越重要颜色越深

```python
sns.palplot(sns.color_palette('Blues'))
```

##### 4.2 如果想要翻转渐变色，在颜色名称后面加 _r 后缀

```python
sns.palplot(sns.color_palette('Blues_r'))
```

#### 5. cubehelix_palette()调色板
色调线性变换

```python
sns.palplot(sns.color_palette("cubehelix",8))
```

```python
sns.palplot(sns.cubehelix_palette(8, start=.5, rot=-.75))
```

```python
sns.palplot(sns.cubehelix_palette(8, start=.75, rot=-.150))
```

#### 6. light_palette()和dark_palette()调用定制连续调色板

```python
sns.palplot(sns.light_palette('green'))
```

```python
sns.palplot(sns.dark_palette('green'))
```

```python
# recerse反转
sns.palplot(sns.light_palette('green',reverse=True))
```

```python
x,y = np.random.multivariate_normal([0,0],[[1,-5],[-5,1]],size=300).T
pal = sns.dark_palette('green',as_cmap=True)
sns.kdeplot(x,y,cmap=pal)
```

```python
sns.palplot(sns.light_palette((210,90,60),input='husl'))
```
