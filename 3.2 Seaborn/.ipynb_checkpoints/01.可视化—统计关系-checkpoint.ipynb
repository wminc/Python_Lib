统计分析是了解数据集中的变量如何相互关联以及这些关系如何依赖于其他变量的过程。用**散点图**和**线图**两种通用的方法来可视化统计关系。
- relplot() 详见seaborn.relplot.md

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

### 一、用散点图关联变量

```python
# 载入自带数据集
tips = pd.read_csv('seaborn-data/tips.csv')
sns.relplot(x="total_bill", y="tip", data=tips)
```

#### 1. 参数hue：对不同组展示不同的颜色

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

#### 2. 参数style：对不同组展示不同的标记样式

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", style="smoker",
            data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", style="time", data=tips)
```

#### 3. 参数size：根据变量的数值，进行标记大小的分组

```python
sns.relplot(x="total_bill", y="tip", hue="size", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", size="size", data=tips)
```

变量的值不用于直接决定点的面积。数据单位中的值范围被规范化为面积单位的范围，这个范围可以自定义

```python
sns.relplot(x="total_bill", y="tip", size="size", sizes=(15, 200), data=tips)
```

### 二、线图的连续性
对于某些数据集，您可能希望了解一个变量中的变化关于时间的函数，或者类似的连续变量。在这种情况下，一个很好的选择是绘制线图。 在 seaborn 中，这可以通过lineplot()函数直接实现，也可以通过设置relplot()的参数kind="line"来实现

```python
df = pd.DataFrame(dict(time=np.arange(500),
                       value=np.random.randn(500).cumsum()))
sns.relplot(x="time", y="value", kind="line", data=df)
```

对于lineplot()，假设您想要绘制y=f(x)的函数，默认行为是在绘制之前按数字x对数据进行排序。但是，这可以被禁用

```python
df = pd.DataFrame(np.random.randn(500, 2).cumsum(axis=0), columns=["x", "y"])
sns.relplot(x="x", y="y", sort=False, kind="line", data=df);
```

#### 1. 聚合和表示不确定性

##### a. 参数ci：聚合方式
更复杂的数据集将对x变量的相同值有多个观测值。<br>
seaborn 的默认行为是通过绘制平均值及 95%的置信区间，在每个x周围聚合多个测量值

```python
fmri = pd.read_csv('seaborn-data/fmri.csv')
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri)
```

置信区间，是使用 bootstrapping 计算的，对于较大的数据集，它可能是时间密集型的。因此，可以禁用它们

```python
sns.relplot(x="timepoint", y="signal", ci=None, kind="line", data=fmri)
```

对于较大的数据，另一个不错的选择是通过绘制标准差，而不是置信区间来表示分布在每个时间点的分布范围

```python
sns.relplot(x="timepoint", y="signal", kind="line", ci="sd", data=fmri)
```

#####  b. 参数estimator：
通过设置estimator参数为None，来完全停用聚合。当数据在每个点上有多个观察值时，这可能会产生奇怪的效果

```python
sns.relplot(x="timepoint", y="signal", estimator=None, kind="line", data=fmri)
```

##### c. 线形图中的参数hue：对不同组展示不同的颜色

```python
sns.relplot(x="timepoint", y="signal", hue="event", kind="line", data=fmri)
```

##### d. 线形图中的参数style：对不同组展示不同的线条样式

```python
sns.relplot(x="timepoint", y="signal", hue="event", style="event",
            kind="line", data=fmri)
```

```python
sns.relplot(x="timepoint", y="signal", hue="region", style="event",
            kind="line", data=fmri)
```

用不同的标记点代替不同的线条样式区分

```python
sns.relplot(x="timepoint", y="signal", hue="region", style="event",
            dashes=False, markers=True, kind="line", data=fmri)
```

##### e.线形图中的参数size：改变线的宽度
虽然size变量通常是数值型的，但是也可以用线宽映射为类别变量。在这样做的时候要小心，因为除了“粗”线和“细”线之外，很难区分更多

```python
sns.relplot(x="timepoint", y="signal", size="event", kind="line", data=fmri)
```

```python
sns.relplot(x="timepoint", y="signal", size="event", sizes=(5,15), kind="line", data=fmri)
```

#### 2. 用日期数据绘图

```python
df = pd.DataFrame(dict(time=pd.date_range("2017-1-1", periods=500),
                       value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df)
# 设置x轴日期旋转一定角度
g.fig.autofmt_xdate()
```

### 三、显示与切面的多种关系
虽然这些函数可以同时显示几个语义变量，但这样做并不总是有效的。您可以创建多个轴并在每个轴上绘制数据的子集
#####  a. 参数col

```python
sns.relplot(x="total_bill", y="tip", hue="smoker",
            col="time", data=tips)
```

#####  b. 参数row

```python
sns.relplot(x="timepoint", y="signal", hue="subject",
            col="region", row="event", height=3,
            kind="line", estimator=None, data=fmri)
```

#####  c. 参数 col_wrap：一行能存在的列数，与row不能同时存在

```python
sns.relplot(x="timepoint", y="signal", hue="event", style="event",
            col="subject", col_wrap=5,
            height=3, aspect=.75, linewidth=2.5,
            kind="line", data=fmri.query("region == 'frontal'"))
```
